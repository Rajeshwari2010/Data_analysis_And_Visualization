```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file=pd.read_csv("Indian_Movies.csv",encoding='latin-1')
file.head()
file.info()
file.describe()
file.isnull().sum()
```

```python

file.dtypes
```

```python
file['Year'].unique()[:50]  # show first 50 unique values
```

```python
#CLEANING THE DATA
file['Year']=file['Year'].str.replace('(','',regex=False).str.replace(')','',regex=False)
file['Duration']=file['Duration'].str.replace('min','')
```

```python
#coverting data types
file['Year']=pd.to_numeric(file['Year'],errors='coerce')
file['Duration']=pd.to_numeric(file['Duration'],errors='coerce')
```

```python
#Check percent of null values in each column
null_percent = (file.isnull().mean() * 100).round(2)
print(null_percent)
```

```python
#drop column with greater null values
file=file.drop(columns=['Votes'])
```

```python
#fill null values for necessary columns
file['Rating']=file['Rating'].fillna(file['Rating'].mean())
```

```python
file['Genre']=file['Genre'].fillna("Other") #fill genre first to use it for duration imputation
file['Duration']=file.groupby('Genre')['Duration'].transform(lambda x: x.fillna(x.median()))
```

```python
#fill remaining columns
file['Year']=file['Year'].fillna(file['Year'].mode()[0])
file['Director']=file['Director'].fillna("Unknown")
file['Actor 1']=file['Actor 1'].fillna("Unknown")
file['Actor 2']=file['Actor 2'].fillna("Unknown")
file['Actor 3']=file['Actor 3'].fillna("Unknown")
```

```python
file.isnull().sum()
```

```python
file['Duration'] = file['Duration'].fillna(file['Duration'].median())
```

```python
file.isnull().sum()
```

```python
file.to_csv("Cleaned_Indian_Movies.csv",index=False)
```
