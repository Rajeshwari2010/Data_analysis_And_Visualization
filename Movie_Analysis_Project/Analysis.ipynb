```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file=pd.read_csv("Cleaned_Indian_Movies.csv")
```

```python
file['Rating']=file['Rating'].round(2)
```

```python
file.head()
file.info()
file.describe()
file.isnull().sum()
```

```python
count=file['Year'].value_counts().sort_index()
plt.figure(figsize=(12,6))
plt.bar(count.index,count.values,color='skyblue')
plt.xlabel('Year')
plt.ylabel("Nummber of movies releasses")
plt.title("Number of movies released each year")
plt.xticks(rotation=45)
plt.show()
```

```python
count=file['Genre'].value_counts().head(20)
plt.figure(figsize=(12,6))
plt.bar(count.index,count.values,color='skyblue')
plt.xlabel('Genre')
plt.ylabel("Count")
plt.title("Top 20 Movie Genres")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.Figure(figsize=(10,5))
plt.hist(file['Rating'],bins=20,color='pink',edgecolor='black')
plt.title("Distribution of movie Ratings")
plt.xlabel("Ratings")
plt.ylabel("Number of movies")
plt.show()
```

```python
sample=file.sample(1000,random_state=1)
plt.Figure(figsize=(10,10))
plt.scatter(sample['Year'],sample['Rating'],color='green',alpha=0.3)
plt.title("Year vs Ratings")
plt.xlabel("Year")
plt.ylabel("Ratings")
plt.show()
```

```python
top_directors=file['Director'].value_counts().head(10)
plt.figure(figsize=(12,5))
bars=plt.barh(top_directors.index,top_directors.values,color='SkyBlue')
plt.bar_label(bars)  
plt.xlabel("Number of movies")
plt.ylabel("Director")
plt.title("Top 10 Directors by number of movies")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(file['Duration'])
plt.title("Boxplot of Movie Durations")
plt.ylabel("Duration (minutes)")
plt.show()
```

```python
plt.Figure(figsize=(10,6))
plt.plot(file['Actor 1'],file['Rating'].head(10),color='Skyblue')
plt.title("Year vs Ratings")
plt.xlabel("Year")
plt.ylabel("Ratings")
plt.show()
```

```python
plt.Figure(figsize=(10,6))
top_Genre=file['Genre'].value_counts().head(5)
plt.pie(top_Genre.values,labels=top_Genre.index,autopct='%1.1f%%',explode=[0.05]*5,startangle=90,colors=sns.color_palette('pastel'))
plt.title("Top 5 Movie Genres")
plt.show()
```

```python
plt.Figure(figsize=(10,5))
top_genres = file['Genre'].value_counts().head(10)
sns.countplot(
    data=file[file['Genre'].isin(top_genres.index)],
    x='Genre',
    order=top_genres.index
)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.Figure(figsize=(10,6))
numeric_df = file.select_dtypes(include='number')

sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm')#works only with numeric data
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(file['Rating'], kde=True)
plt.title("Distribution of Ratings")
plt.show()
```

```python
# 1. Find top 10 genres
top10_genres = file['Genre'].value_counts().head(10).index

# 2. Filter dataframe
filtered_df = file[file['Genre'].isin(top10_genres)]

# 3. Plot violin chart
plt.figure(figsize=(12,6))
sns.violinplot(data=filtered_df, x='Genre', y='Duration')
plt.xticks(rotation=90)
plt.title("Duration Distribution for Top 10 Movie Genres")
plt.show()
```
